```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = pd.read_csv("processed_data.csv")
```

```python
print("We have {} rows.".format(data.shape[0]))
print("We have {} columns".format(data.shape[1]),"\n")
```

```python
print("Total Null Values in Dataset:", np.sum(pd.isnull(data)), "\n")
```

```python
print("Unique values in 'city' column:")
print(data['city'].unique(), "\n")
```

```python
for col in data.columns:
    mode_value = data[col].mode()[0]
    data[col] = data[col].fillna(mode_value)

```

```python
for col in data.columns:
    try:
        data[col] = data[col].astype(np.int64)
    except:
        pass

print("Numeric columns converted to INT where possible.\n")
```

```python
print("Datatypes of All Columns:")
print(data.dtypes, "\n")
```

```python
x = data.iloc[:, 1:]  # except price
y = data.iloc[:, 0]   # price column

print("X shape:", x.shape)
print("Y shape:", y.shape, "\n")
```

```python
cat_columns = x.select_dtypes(['object']).columns

if len(cat_columns) > 0:
    x[cat_columns] = x[cat_columns].apply(lambda col: pd.factorize(col)[0])
    print("Object columns converted to integer encoding.\n")
else:
    print("No object columns found to convert.\n")

print("Final Pre-Processed Dataset Ready!")
```
