```python
class Graph:
    def __init__(self, connections):
        self.connections = connections 
    def get_neighbors(self, vertex):
        return self.connections[vertex]
    def heuristic(self, vertex):
        heuristic_estimates = {
            'A': 1,
            'B': 1,
            'C': 1,
            'D': 1
        }
        return heuristic_estimates[vertex]

    def a_star_search(self, start_node, goal_node):
        open_list = [start_node]
        closed_list = []
        g_cost = {start_node: 0}
        previous_node = {start_node: start_node}

        while len(open_list) > 0:
            current_node = None
            for node in open_list:
                if current_node is None or g_cost[node] + self.heuristic(node) < g_cost[current_node] + self.heuristic(current_node):
                    current_node = node

            if current_node is None:
                print("Path does not exist!")
                return None

            if current_node == goal_node:
                path = []
                while previous_node[current_node] != current_node:
                    path.append(current_node)
                    current_node = previous_node[current_node]
                path.append(start_node)
                path.reverse()

                print("Path found:", path)
                return path
            for (neighbor, cost) in self.get_neighbors(current_node):
                if neighbor not in open_list and neighbor not in closed_list:
                    open_list.append(neighbor)
                    previous_node[neighbor] = current_node
                    g_cost[neighbor] = g_cost[current_node] + cost
                else:
                    if g_cost[neighbor] > g_cost[current_node] + cost:
                        g_cost[neighbor] = g_cost[current_node] + cost
                        previous_node[neighbor] = current_node

                        if neighbor in closed_list:
                            closed_list.remove(neighbor)
                            open_list.append(neighbor)

            open_list.remove(current_node)
            closed_list.append(current_node)

        print("Path does not exist!")
        return None

connections = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('D', 5)],
    'C': [('D', 12)]
}

graph = Graph(connections)
graph.a_star_search('A', 'D')
```
