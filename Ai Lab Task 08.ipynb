```python
def minimax(depth, isMaximizing, scores, maxDepth):
    if depth == maxDepth:
        return scores[depth]

    if isMaximizing:
        best = float('-inf')
        best = max(best, minimax(depth + 1, False, scores, maxDepth))
        return best
    else:
        best = float('inf')
        best = min(best, minimax(depth + 1, True, scores, maxDepth))
        return best


scores = [3, 5, 2, 9]

maxDepth = len(scores) - 1

result = minimax(0, True, scores, maxDepth)
print("Optimal value:", result)
```
